```python
title = 'Uptown Funk'
genre = 'hip-hop'
```

```python
selected_song = pd.DataFrame(list(db['total'].find({'title':title}))).reset_index(drop = True).set_index('title')
selected_genre_songs = pd.DataFrame(list(db[genre].find({}))).reset_index(drop = True).set_index('title')
```

```python
import numpy as np
selected_song = selected_song.select_dtypes([np.number])
selected_genre_songs = selected_genre_songs.select_dtypes([np.number])
```

```python
c = pd.concat([selected_song,selected_genre_songs]).dropna(axis = 1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarty = cosine_similarity(c)
```

```python
sim_df = pd.DataFrame(similarty, index = c.index, columns = c.index)
result = sim_df['Uptown Funk'].sort_values(ascending = False).drop('Uptown Funk').to_frame()

result = pd.DataFrame(result.index)
result
```

```python
def get_recommend(title, genre, n=5):
    conn = pymongo.MongoClient('localhost', 27017)
    db = conn['DA']
    if genre in db.list_collection_names():
        selected_song = pd.DataFrame(list(db['total'].find({'title':title}))).reset_index(drop = True).set_index('title')
        selected_genre_songs = pd.DataFrame(list(db[genre].find({}))).reset_index(drop = True).set_index('title')
        selected_song = selected_song.select_dtypes([np.number])
        selected_genre_songs = selected_genre_songs.select_dtypes([np.number])
    else: pass
    
    df = pd.concat([selected_song, selected_genre_songs], axis = 0).dropna(axis = 1)
    
    similarty = cosine_similarity(df)
    sim_df = pd.DataFrame(similarty, index = df.index, columns = df.index)
    
    result = sim_df[title].sort_values(ascending = False).drop(title).to_frame()
    
    return pd.DataFrame(result.index).head(n)
```

```python
get_recommend(title, genre)
```
