```python
#sudo apt-get install python3-pandas
import pandas as pd
import numpy as np
import re
import os
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import librosa.display

def split_wav(data, sample_rate, start, end):
    start *= sample_rate
    end *= sample_rate
    return data[start:end]
```

```python
path = '/home/ubuntu/DA'

file_list = os.listdir(path)
file_list

os.chdir('/home/ubuntu/DA/dataframe')
```

```python
balad_file_list = os.listdir(path+ '/' + 'balad')
balad_file_list.remove('.ipynb_checkpoints')
balad_file_list
```

```python
#빈 데이터프레임 생성
balad_df = pd.DataFrame(columns = ['song', 'genre'])
balad_df

#음원 데이터프레임 생성


for i in range(len(balad_file_list)):
    balad_df = balad_df.append(pd.DataFrame([[balad_file_list[i], 'balad']],
                                              columns = ['song', 'genre']),
                                 ignore_index = True)
        
print(balad_df)    
```

```python
#데이터프레임 갱신
for i in range(0,len(balad_file_list)):
        if balad_df.iloc[i,0] in balad_file_list == True:
            continue
        else: balad_df.append(pd.DataFrame([[balad_file_list[i], 'balad']],
                                           columns = ['song', 'genre']),
                              ignore_index = True)
print(balad_df)        
```

```python
balad_df.info()
```

```python
#title, genre 컬럼 생성
for i in range(len(balad_df)):
    balad_df['song'].iloc[i] = balad_df['song'].iloc[i].replace(' _ ', ' - ', 1)
    balad_df['song'].iloc[i] = balad_df['song'].iloc[i].replace(' _', ' - ', 1)
    balad_df['song'].iloc[i] = balad_df['song'].iloc[i].replace(' — ', ' - ', 1)
    balad_df['song'].iloc[i] = balad_df['song'].iloc[i].replace('- ', ' - ', 1)
    balad_df['song'].iloc[i] = balad_df['song'].iloc[i].replace(' -', ' - ', 1)
    balad_df['song'].iloc[i] = balad_df['song'].iloc[i].replace('-', ' - ', 1)
    balad_df['song'].iloc[i] = balad_df['song'].iloc[i].replace('[MV]', '', 1)



balad_df['singer'] = balad_df['song'].str.split(' - ').str[0]
print(balad_df['singer'], end = '\n--------------------------------------------\n')

balad_df['title'] = balad_df['song'].str.split(' - ').str[1]

for i in range(len(balad_df)):
    if balad_df['title'].str.contains('Feat|feat|Prod')[i] == True:
        continue
    else:
        balad_df['title'].iloc[i] = re.sub(r'\([^)]*\)', '', balad_df['title'].iloc[i])




balad_df['title'] = balad_df['title'].str.split('.mp3').str[0] 
balad_df['title'] = balad_df['title'].str.split('-').str[0]
balad_df['title'] = balad_df['title'].str.split('가사').str[0]
balad_df['title'] = balad_df['title'].str.split('[').str[0] 
balad_df['title'] = balad_df['title'].str.split('Official').str[0] 

print(balad_df['title'])
```

```python
#BPM 추출
bpm = []

for i in range(len(balad_file_list)):
    y, sr = librosa.load('/home/ubuntu/DA/balad/'+ balad_file_list[i])
    y = split_wav(y, sr, 0, 60)
    tempo, _ = librosa.beat.beat_track(y, sr)
    bpm.append(tempo)

balad_df['tempo1'] = pd.DataFrame(bpm)
```

```python
#BPM 추출
bpm = []

for i in range(len(balad_file_list)):
    y, sr = librosa.load('/home/ubuntu/DA/balad/'+ balad_file_list[i])
    y = split_wav(y, sr, 60, 120)
    tempo, _ = librosa.beat.beat_track(y, sr)
    bpm.append(tempo)

balad_df['tempo2'] = pd.DataFrame(bpm)
```

```python
#FT (푸리에 변환)
#시간 영역 데이터를 주파수 영역으로 변경
#y축: 주파수(로그 스케일)
#푸리에 변환을 한 후 곡 별산술통계량을 내어보자.

mean_ft = []
var_ft = []

for i in range(len(balad_file_list)):
    y, sr = librosa.load('/home/ubuntu/DA/balad/'+ balad_file_list[i])
    y = split_wav(y, sr, 0, 60)
    D = np.abs(librosa.stft(y, n_fft = 2048, hop_length = 512))
    mean_D = np.mean(D)
    mean_ft.append(mean_D)
    var_D = np.var(D)
    var_ft.append(var_D)

balad_df['mean_ft1'] = pd.DataFrame(mean_ft)
balad_df['var_ft1'] = pd.DataFrame(var_ft)
```

```python
#FT (푸리에 변환)
#시간 영역 데이터를 주파수 영역으로 변경
#y축: 주파수(로그 스케일)
#푸리에 변환을 한 후 곡 별산술통계량을 내어보자.

mean_ft = []
var_ft = []

for i in range(len(balad_file_list)):
    y, sr = librosa.load('/home/ubuntu/DA/balad/'+ balad_file_list[i])
    y = split_wav(y, sr, 60, 120)
    D = np.abs(librosa.stft(y, n_fft = 2048, hop_length = 512))
    mean_D = np.mean(D)
    mean_ft.append(mean_D)
    var_D = np.var(D)
    var_ft.append(var_D)

balad_df['mean_ft2'] = pd.DataFrame(mean_ft)
balad_df['var_ft2'] = pd.DataFrame(var_ft)
```

```python
mean_mel = []
var_mel = []

for i in range(len(balad_file_list)):
    y, sr = librosa.load('/home/ubuntu/DA/balad/'+ balad_file_list[i])
    y = split_wav(y, sr, 0, 60)
    s = librosa.feature.melspectrogram(y, sr)
    s_DB = librosa.amplitude_to_db(s, ref = np.max)
    mean_s_DB = np.mean(s_DB)
    mean_mel.append(mean_s_DB)
    var_s_DB = np.var(s_DB)
    var_mel.append(var_s_DB)

balad_df['mean_mel1'] = pd.DataFrame(mean_mel)
balad_df['var_mel1'] = pd.DataFrame(var_mel)
```

```python
#Harmonic and Percussive components
#Harmonics: 사람의 귀로 구분할 수 없는 특징들 (음악의 색깔)
#음악의 색깔은 평균 추출
#Percussives: 리듬과 감정을 나타내는 충격파
#충격파는 min과 max를 추출
mean_h = []
var_h = []
mean_p = []
var_p = []
for i in range(len(balad_file_list)):
    y, sr = librosa.load('/home/ubuntu/DA/balad/'+ balad_file_list[i])
    y = split_wav(y, sr, 0, 60)
    harm, perc = librosa.effects.hpss(y)
    mean_h.append(np.mean(harm))
    var_h.append(np.var(harm))
    mean_p.append(np.mean(perc))
    var_p.append(np.var(perc))

balad_df['mean_harm1'] = pd.DataFrame(mean_h)
balad_df['var_harm1'] = pd.DataFrame(var_h)
balad_df['mean_perc1'] = pd.DataFrame(mean_p)
balad_df['var_perc1'] = pd.DataFrame(var_p)
```
